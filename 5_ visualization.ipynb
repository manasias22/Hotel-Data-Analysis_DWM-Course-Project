```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_excel("dwm1.xlsx")
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
plt.scatter(dataset['Price(₹)'],dataset['Discount (% OFF)'], color='blue')
plt.title("scatter plot")
plt.xlabel("Prize")
plt.ylabel("Discount")
plt.show()
```

```python
plt.hist(dataset['Rating(10)'], bins=20)
plt.title("Histogram")
plt.xlabel("Rating(10)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(dataset['Price(₹)'], bins=25)
plt.title("Histogram")
plt.xlabel("Price(₹)")
plt.ylabel("Frequency")
plt.show()
```

```python
sb.boxplot(x="Rating(10)", data=dataset)
plt.title("Box Plot")
```

```python
sb.boxplot(x="Price(₹)", data=dataset)
plt.title("Box Plot")
```
